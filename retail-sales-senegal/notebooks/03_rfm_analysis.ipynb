# 03 - Analyse RFM (Récence, Fréquence, Montant)
Segmentation des clients pour identifier les Champions, clients fidèles, à risque, etc.

```python
import sys
sys.path.append('..')
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from src.rfm import compute_rfm
from src.config import PROCESSED_DATA_PATH, FIGURES_DIR

sns.set_theme(style='whitegrid')
df = pd.read_csv(PROCESSED_DATA_PATH, parse_dates=['InvoiceDate'])
rfm = compute_rfm(df)
rfm.head()
```

## 1. Distribution des segments clients

```python
segment_counts = rfm['Segment'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=segment_counts.values, y=segment_counts.index, palette='viridis')
plt.title('Répartition des clients par segment RFM')
plt.xlabel('Nombre de clients')
plt.tight_layout()
plt.savefig(FIGURES_DIR / 'segments_rfm.png', dpi=150)
plt.show()
```

## 2. Valeur moyenne par segment

```python
segment_value = rfm.groupby('Segment')['Monetary'].mean().sort_values(ascending=False)
print(segment_value)
```

## 3. Export final pour Power BI

```python
rfm.to_csv('../data/processed/rfm_segments.csv', index=False)
print('Fichier RFM exporté pour Power BI.')
```
