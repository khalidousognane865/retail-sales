# 02 - Exploratory Data Analysis (EDA)
Analyse exploratoire des ventes : KPIs, tendances temporelles, top produits.

```python
import sys
sys.path.append('..')
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from src.config import PROCESSED_DATA_PATH, FIGURES_DIR

sns.set_theme(style='whitegrid')
df = pd.read_csv(PROCESSED_DATA_PATH, parse_dates=['InvoiceDate'])
df.head()
```

## 1. KPIs globaux

```python
ca_total = df['TotalPrice'].sum()
panier_moyen = df.groupby('InvoiceNo')['TotalPrice'].sum().mean()
nb_clients = df['CustomerID'].nunique()
nb_commandes = df['InvoiceNo'].nunique()

print(f'CA Total : {ca_total:,.2f}')
print(f'Panier moyen : {panier_moyen:,.2f}')
print(f'Clients uniques : {nb_clients:,}')
print(f'Commandes : {nb_commandes:,}')
```

## 2. Évolution mensuelle du CA

```python
ca_mensuel = df.groupby('YearMonth')['TotalPrice'].sum().reset_index()
plt.figure(figsize=(12, 5))
sns.lineplot(data=ca_mensuel, x='YearMonth', y='TotalPrice', marker='o', color='#2E75B6')
plt.xticks(rotation=45)
plt.title('Évolution mensuelle du chiffre d\'affaires')
plt.ylabel('CA')
plt.xlabel('Mois')
plt.tight_layout()
plt.savefig(FIGURES_DIR / 'ca_mensuel.png', dpi=150)
plt.show()
```

## 3. Top 10 produits par chiffre d'affaires

```python
top_produits = df.groupby('Description')['TotalPrice'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_produits.values, y=top_produits.index, palette='Blues_r')
plt.title('Top 10 produits par chiffre d\'affaires')
plt.xlabel('CA')
plt.tight_layout()
plt.savefig(FIGURES_DIR / 'top_produits.png', dpi=150)
plt.show()
```

## 4. Heatmap des ventes par jour de semaine et heure

```python
heatmap_data = df.pivot_table(index='DayOfWeek', columns='Hour', values='TotalPrice', aggfunc='sum', fill_value=0)
ordre_jours = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
heatmap_data = heatmap_data.reindex(ordre_jours)

plt.figure(figsize=(14, 6))
sns.heatmap(heatmap_data, cmap='YlOrRd', linewidths=0.5)
plt.title('Heatmap des ventes : jour de semaine x heure')
plt.tight_layout()
plt.savefig(FIGURES_DIR / 'heatmap_ventes.png', dpi=150)
plt.show()
```

## 5. Répartition par pays

```python
ventes_pays = df.groupby('Country')['TotalPrice'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=ventes_pays.values, y=ventes_pays.index, palette='Greens_r')
plt.title('Top 10 pays par chiffre d\'affaires')
plt.tight_layout()
plt.savefig(FIGURES_DIR / 'ventes_par_pays.png', dpi=150)
plt.show()
```
